### COMMON ISSUES / TROUBLESHOOTING

#### Issue 1: Virtual Machine is Slow or Unresponsive
✔ **Solution:** Increase allocated RAM and CPU cores in VMware settings.

#### Issue 2: MySQL Fails to Start
✔ **Solution:** Open Terminal and run:

```sh
sudo service mysql restart
```

#### Issue 3: Toolkit Launcher Doesn't Open
✔ Solution: Manually start the toolkit from Terminal:

Bash

cd ~/ailtk_launcher
conda activate ailtk-env
voila

#### Issue 4: Google Gemini API or Key Not Working
✔ Solution: Ensure the key is active in Google Cloud Console. Refer to the Google Gemini Developer API.

#### Issue 5: Running Python in VSCode results in an error stating “Missing Dependency”
✔ Solution: Install the missing package using Conda or Pip depending on the package. Also, refer to the requirements.yml file in the root folder of the toolkit (~/capstone-ailtk).

Using Conda:
```sh

Bash

conda activate ailtk-env
conda install <missing-package-name>
```

Using Pip:
```sh

Bash

conda activate ailtk-env
pip install <missing-package-name>
```


Updating from requirements.yml:

```sh

cd ~/capstone-ailtk
conda activate ailtk-env
conda env update --name ailtk-env --file requirements.yml --prune
```

#### Issue 6: Conda Virtual Environment (ailtk-env) returns an error in VSCode and fails to start.
✔ Solution: Restart the terminal instance and enter:

```sh
Bash

conda init
conda activate ailtk-env
```

### FAQs
Q1: How long does the toolkit take to load?
A: The initial load may take 3-5 minutes due to caching and resource allocation. Subsequent loads are faster.

Q2: Can I use my existing Python environment?
A: No. Use the pre-configured environment inside the virtual machine to avoid compatibility issues.

Q3: What if my VMware doesn't support the virtual machine?
A: Ensure virtualization is enabled in BIOS and you are using a compatible VMware version.

Q4: How do I reset the toolkit if I encounter persistent issues?
A: Reboot the virtual machine and delete temporary files in the /tmp/ directory.

Q5: Is there a way to access the toolkit without VMware?
A: Currently, the toolkit is only available through the virtual machine. A web-based version is in development.

Q6: Where can I find additional learning resources?
A: Check the Viola Home tab and the Resources folder on the virtual machine’s desktop.
